```python
# [Problema] dato un rettangolo di perimetro 100, quali sono le
# lunghezze dei lati l e h che massimizzano l’area A del rettangolo?
#P = 100

# Analitical solution
# l = P/2 - h = 50 - h
# A = l*h = (50 - h)h = 50h - h^2
# Max A where A' = 50 - 2h = 0
# h = 25
# l = 50 - 25 = 25

# Procedural solution
# Per ogni valore di l compreso tra 1 e 50 (il semi-perimetro)
# calcola h = 50 – l
# calcola l’area del rettangolo avente per lati l e h
# conserva i valori l e h del rettangolo con area maggiore
def maximize_area(P):
    max_area = 0
    best_h = 0
    best_l = 0
    
    for h in range(1, P // 2):
        l = P // 2 - h
        area = l * h
        if area > max_area:
            max_area = area
            best_h = h
            best_l = l
    
    return best_l, best_h


perimeter = 100
length, height = maximize_area(perimeter)
print(f"Length: {length}, Height: {height}")
```

```python
import matplotlib.pyplot as plt

def maximize_area_with_plot(P):
    max_area = 0
    best_h = 0
    best_l = 0
    areas = []
    hs = []
    ls = []
    
    for h in range(1, P // 2 + 1):
        l = P // 2 - h
        area = l * h
        areas.append(area)
        hs.append(h)
        ls.append(l)
        
        if area > max_area:
            max_area = area
            best_h = h
            best_l = l
        
        if h % 25 == 0:
            plt.plot(hs, areas, label=f'Iteration {h}')
            plt.xlabel('Height (h)')
            plt.ylabel('Area')
            plt.title('Area vs Height')
            plt.legend()
            plt.show()
    
    return best_l, best_h

# Example usage
perimeter = 100
length, height = maximize_area_with_plot(perimeter)
print(f"Length: {length}, Height: {height}")
```
