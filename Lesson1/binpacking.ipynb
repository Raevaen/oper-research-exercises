Un dispositivo di memoria è disponibile in blocchi di dimensione data (per es. 10 Mb).
Un insieme di file deve essere memorizzato tenendo conto che ogni file ha una dimensione data e non può essere frammentato in più blocchi.
determinare una soluzione numerica (euristica).

```python
# Dati forniti
block_size = 10000  # Dimensione del blocco in kb
files_data = [ # Lista di tuple (file_size, number_of_files)
    (3834, 10), (6928, 11), (1506, 22), (204, 89), (3408, 83), 
    (6623, 21), (6102, 75), (1755, 67), (5653, 18), (4809, 15),
    (4286, 2), (3465, 30), (965, 75), (4741, 79), (3550, 16),
    (3658, 28), (6753, 4), (1090, 29), (5245, 6), (6131, 33)
]

files_data = sorted(files_data, key=lambda x: x[0], reverse=True)  # Ordina i file per dimensione decrescente

# Espande i dati in una lista completa di file
files = []
for size, count in files_data:
    files.extend([size] * count)

# Algoritmo First-Fit Decreasing (FFD)
blocks = []  # Lista di blocchi (ogni blocco è una lista di file)

for file_size in files:
    # Tento di inserire il file nel primo blocco disponibile
    placed = False
    for block in blocks:
        if sum(block) + file_size <= block_size:
            block.append(file_size)
            placed = True
            break
    # Se non c'è spazio nei blocchi esistenti, creo un nuovo blocco
    if not placed:
        blocks.append([file_size])
# Numero totale di blocchi necessari
num_blocks = len(blocks)

# Restituzione dei risultati
print("Number of blocks:", num_blocks)
print("Blocks:", blocks)
```

```python
# Dati forniti
block_size = 10000  # Dimensione del blocco in kb
files_data = [ # Lista di tuple (file_size, number_of_files)
    (3834, 10), (6928, 11), (1506, 22), (204, 89), (3408, 83), 
    (6623, 21), (6102, 75), (1755, 67), (5653, 18), (4809, 15),
    (4286, 2), (3465, 30), (965, 75), (4741, 79), (3550, 16),
    (3658, 28), (6753, 4), (1090, 29), (5245, 6), (6131, 33)
]

files_data = sorted(files_data, key=lambda x: x[0], reverse=True)  # Ordina i file per dimensione decrescente

# Espande i dati in una lista completa di file
files = []
for size, count in files_data:
    files.extend([size] * count)

# Algoritmo 2 by Ema, non so se ha un nome
midpoint = len(files) // 2

# Divido la lista di files in due parti
first_half = files[:midpoint]  # Ordine decrescente
second_half = sorted(files[midpoint:]) # Ordine crescente


def place_smallest(block):
    for file_size in second_half:
        # Tenta di inserire il file nel blocco disponibile
        if sum(block) + file_size <= block_size:
            block.append(file_size)
            second_half.remove(file_size)
        else:
            return

blocks = []  # Lista di blocchi (ogni blocco è una lista di file)
for file_size in first_half:
    # Tenta di inserire il file nel primo blocco disponibile
    placed = False
    for block in blocks:
        if sum(block) + file_size <= block_size:
            block.append(file_size)
            placed = True
            place_smallest(block)
            break
    # Se non c'è spazio nei blocchi esistenti, crea un nuovo blocco
    if not placed:
        blocks.append([file_size])



num_blocks = len(blocks)

# Restituzione dei risultati
print("Number of blocks:", num_blocks)
print("Blocks:", blocks)
```
